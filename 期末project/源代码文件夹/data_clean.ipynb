# 词云图预处理

```python
import jieba
```

```python
stopwords=[]
for word in open('stopwords.txt','r',encoding = 'utf-8'):
    stopwords.append(word.replace('\n',''))
```

```python
simanan_dic = {}
```

```python
for data in simanan:
    text = data[0]['text']
    text_list = jieba.cut(text, cut_all=False)
    for ele in text_list:
        if ele not in stopwords:
            if ele in simanan_dic.keys():
                simanan_dic[ele] = simanan_dic[ele]+1
            else:
                simanan_dic[ele] = 1
    
```

```python
simanan_sort = sorted(simanan_dic.items(),key=lambda x:x[1],reverse=True)
```

```python
with open('simanan_sort.txt','w',encoding = 'utf-8')as f1:
    f1.write('word')
    f1.write(',')
    f1.write('freq')
    f1.write('\n')
    for key in simanan_sort[:100]:
        f1.write(key[0])
        f1.write(',')
        f1.write(str(key[1]))
        f1.write('\n')
    
f1.close()
```

# 认证职业/性别/地点

```python
simanan_verified = {}
simanan_gender = {}
simanan_location = []#'province','city'
simanan_user = {}
simanan_id = []
```

```python
for data in simanan:
    for person in data[1:]:
        uid = person["uid"]
        if uid in simanan_id:
            continue
        else:
            simanan_id.append(uid)
        text = person['user_description']
        verified_reason = person['verified_reason'].replace(' ','，').split('，')
        gender = person['gender']
        province = person['province']
        city = person['city']

        for ele in verified_reason:
            if ele in simanan_verified.keys():
                simanan_verified[ele] = simanan_verified[ele]+1
            else:
                simanan_verified[ele] = 1
    
        text_list = jieba.cut(text, cut_all=False)
        for ele in text_list:
            if ele not in stopwords:
                if ele in simanan_user.keys():
                    simanan_user[ele] = simanan_user[ele]+1
                else:
                    simanan_user[ele] = 1

        if gender in simanan_gender.keys():
                simanan_gender[gender] = simanan_gender[gender]+1
        else:
            simanan_gender[gender] = 1

        simanan_location.append([province, city])
```

# 城市对照

```python
import csv
```

```python
index = {}
```

```python
csv_reader = csv.reader(open("pro.csv",'r',encoding='utf-8'))
for row in csv_reader:
    index[int(row[0])] = row[1]
```

```python
simanan_pro = {}
for data in simanan_location:
    pro_index = data[0]
    if pro_index not in index:
        continue
    pro = index[pro_index]
    if pro not in simanan_pro.keys():
        simanan_pro[pro] = 1
    else:
        simanan_pro[pro] = simanan_pro[pro] + 1
simanan_pro_sort = sorted(simanan_pro.items(),key=lambda x:x[1],reverse=True)
```

```python
duzijian_pro = {}
for data in duzijian_location:
    pro_index = data[0]
    if pro_index not in index:
        continue
    pro = index[pro_index]
    if pro not in duzijian_pro.keys():
        duzijian_pro[pro] = 1
    else:
        duzijian_pro[pro] = duzijian_pro[pro] + 1
duzijian_pro_sort = sorted(duzijian_pro.items(),key=lambda x:x[1],reverse=True)
    
```
